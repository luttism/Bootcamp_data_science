1) Obteve os dados do banco de dados "Datasus" gerado no Tabnet - Gastos de Ambiente hospitalar pelo Governo Brasileiro - a partir de 2008. Os dados foram exportados no format "csv" (comma-separated-values).

2) Importar os dados usando a biblioteca do Python chamada "pandas" com alias "pd":

```
import pandas as pd

dados = pd.read_csv("/content/A160324189_28_143_208.csv", encoding="ISO-8859-1", 
                    skiprows = 3, sep=";", skipfooter=12, thousands=".", decimal=",")
#mostra as 5 primeiras linhas do dataframe "dados"
dados.head() 
```

3) Formatar os dados para duas casas decimais, pois, por "default" é mostrado em notação científica.

```
pd.options.display.float_format = "{:.2f}".format
```

4) Gerando a tabela de dados usáveis, de acordo com a aula 02.


```
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
colunas_usaveis
```

```
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```
usaveis = usaveis.set_index("Unidade da Federação")
usaveis
```

Retirando a coluna "Total":

```
usaveis = usaveis.drop("Total", axis=1)
usaveis.head()
```

Gerou o gráfico sem formatação:

```
usaveis.T.plot(figsize=(10,6))
```

**5) Aula 02 - Desafio 01: Reposicionar a legenda do gráfico em uma posição mais adequada**

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = usaveis.T.plot(figsize=(10,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.xlabel('Ano/Mês')
plt.ylabel('Valores')
plt.grid(True)
plt.title("Gastos Hospitalares no Brasil (R$) - por Unidades da Federação")
plt.legend(usaveis.index, loc="best", bbox_to_anchor=(1,1))
```

**6) Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferência**

```
# Estados de Escolha: Região Nordeste do Brasil - Maranhão, Piauí, Ceará, Bahia, Pernambuco
estados_nordeste = usaveis.T[["21 Maranhão", "22 Piauí", "23 Ceará", "26 Pernambuco", "29 Bahia"]]

estados_nordeste.head()
```

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = estados_nordeste.plot(figsize=(10,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.xlabel('Ano/Mês')
plt.ylabel('Valores')
plt.grid(True)
plt.title("Gastos Hospitalares no Brasil (R$) - Cinco Estados da Região Nordeste")
plt.legend(estados_nordeste.columns, loc="best", bbox_to_anchor=(1,1))
```
